```python
# a. https://leetcode.com/problems/remove-outermost-parentheses/

class Stack:
    def __init__(self):
        self.top = -1
        self.arr = []
        
    def append(self, value):
        self.top += 1
        self.arr.append(value)
        
    def pop(self):
        if(self.top == -1):
            return None
        self.top -= 1
        return self.arr[self.top+1]
        
class Solution:
    def removeOuterParentheses(self, S: str) -> str:
        s = Stack()
        result = ""
        initial = 0
        for i,x in enumerate(S):
            if(x == '('):
                s.append(x)
            else:
                pop = s.pop()
                if(s.top == -1):
                    result += S[initial+1:i]
                    initial = i+1
                    
        return result
```

```python
# c. https://leetcode.com/problems/remove-all-adjacent-duplicates-in-string/

class Solution:
    def removeDuplicates(self, S: str) -> str:
        result = []
        for x in S:
            if(len(result) == 0 or result[-1] != x):
                result.append(x)
            else:
                result.pop()
        return "".join(result)
```

```python
# d. https://leetcode.com/problems/design-a-stack-with-increment-operation/

class CustomStack:

    def __init__(self, maxSize: int):
        self.maxSize = maxSize
        self.arr = [None for _ in range(maxSize)]
        self.top = -1

    def push(self, x: int) -> None:
        if(self.top < self.maxSize-1):
            self.top += 1
            self.arr[self.top] = x

    def pop(self) -> int:
        if(self.top == -1):
            return -1
        popped = self.arr[self.top]
        self.top -= 1
        return popped

    def increment(self, k: int, val: int) -> None:
        x,y = 0,0
        while(x < k and y <= self.top):
            self.arr[x] += val
            x += 1
            y += 1


# Your CustomStack object will be instantiated and called as such:
# obj = CustomStack(maxSize)
# obj.push(x)
# param_2 = obj.pop()
# obj.increment(k,val)
```

```python
# e. https://leetcode.com/problems/valid-parentheses/

class Solution:
    def isValid(self, s: str) -> bool:
        d = {")":"(","]":"[","}":"{"}
        result = []
        for x in s:
            if(len(result) == 0 or x not in d.keys()):
                result.append(x)
            elif(result[-1] == d[x]):
                result.pop()
            else:
                result.append(x)
        if(len(result) == 0):
            return True
        else:
            return False
```

```python
# f. https://leetcode.com/problems/minimum-add-to-make-parentheses-valid/

class Solution:
    def minAddToMakeValid(self, S: str) -> int:
        result = []
        for x  in S:
            if(len(result) == 0):
                result.append(x)
            elif(x == ")" and result[-1] == "("):
                result.pop()
            else:
                result.append(x)
        return len(result)
```
